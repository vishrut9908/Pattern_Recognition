## Part 4

We are going to use `identity_matrix` to scramble the rows and columns of `img_average_channel`.

1. First, create a numpy array that contains the numbers $0,1, \cdots, N$ where $N$ is the number of rows in the image.

2. Use `numpy.random.shuffle` to shuffle the indices. Call the shuffled array, `rows`

4. Print `rows` and inspect the elements.

```python
# Creating an array row
number_of_rows = img_average_channel.shape[0]
rows = np.arange(number_of_rows)

# Shuffling the indices of rows
np.random.shuffle(rows)

# Printing and inspecting the shuffled array
print(rows)
```

## Part 5

1. Use your array of random row indices to create a new matrix `T` from `identity_matrix`. Overall, `T` has the same elements as `identity_matrix`, but the order of the rows of `T` should be according to the variable `rows`
2. Take a moment to think about what we are doing here. Describe in your own words what matrix `T` is.
3. Sanity check, print the shape of `T` and make sure it is the same size as `identity_matrix` and `img_average_channel`. If you need to,  reshape it to have the same size.


```python
# Creating the matrix T
T = identity_matrix[rows]

print("""Matrix T represents a permutation of the rows of identity_matrix.
This means that T is a modified version of the identity matrix where the rows have been rearranged according to the shuffled indices stored in the rows array.
Applying T to the image would lead to a distortion where the rows of the image are rearranged.
This can result to a stretching or compression effect vertically.""" + "\n")

# Shape of matrix T
height_of_T, width_of_T = img_average_channel.shape
print("The dimensions of matrix T are " + f"Width: {width_of_T}, Height: {height_of_T}" + "\n")
```

## Part 6

1. Matrix multiply `T` and `img_average_channel` and store the result in `transformed_image`. Check the shape of the result.
2. Plot `transformed_image` and describe the pattern you see.
3. Think about what each dot product in the matrix multiplication is doing. Describe in your own words what this transformation is doing to the original image.

```python
# Creating transformed_image
transformed_image = np.matmul(T, img_average_channel)
print("The dimensions of matrix transformed image are " + f"Width: {transformed_image.shape[1]}, Height: {transformed_image.shape[0]}" + "\n")

# Plotting the transformed image
plt.imshow(transformed_image)
plt.show()
print("\n")

print("""Depending on how the rows of T have been shuffled, this operation can have multiple effects on the original image:
Stretching: If the rows are rearranged in a specific way, it can result in a stretching effect on the image vertically, where some rows may be duplicated or omitted.
Vertical Displacement: Certain rows may be moved up or down in relation to their original position, leading to a displacement effect.""")
```

## Part 7

1. We now want to retrieve the original image. Is matrix `T` invertible? How would you show that it is invertible?
2. Compute the inverse and store the matrix in `Tinv`.
3. Perform the inverse of the scrambling operation and plot the result to show that you retrieved the original image.

```python
# Checking the invertibility of matrix T
height, width = T.shape
determinant = np.linalg.det(T)
if determinant != 0 and height == width:
    print("Matrix T is inversible \n")
else:
    print("Matrix T is not inversible \n")

# Computing the matrix in Tinv
Tinv = np.linalg.inv(T)

# Retrieving the original image
retrieved_image = np.matmul(Tinv, transformed_image)
plt.imshow(retrieved_image)
plt.show()
```

## Part 8: Extra Points

```python
number_of_columns = img_average_channel.shape[1]
columns = np.arange(number_of_columns)
np.random.shuffle(columns)
T = identity_matrix[:, columns]
transformed_image = np.matmul(img_average_channel, T)

plt.imshow(transformed_image)
plt.show()
print("\n")

height, width = T.shape
determinant = np.linalg.det(T)
if determinant != 0 and height == width:
    print("Matrix T is inversible \n")
else:
    print("Matrix T is not inversible \n")

Tinv = np.linalg.inv(T)
retrieved_image = np.matmul(transformed_image, Tinv)
plt.imshow(retrieved_image)
plt.show()
```

## Extra Points

* Now can you perform the scrambling on __both the columns and rows__ of the original image?  Write the code and plot the scrambled image. Then invert the operations and show that you can recover the original image.

```python
number_of_rows, number_of_columns = img_average_channel.shape  # Get the dimensions
identity_matrix_rows = np.eye(number_of_rows)
rows = np.arange(number_of_rows)
np.random.shuffle(rows)
T_rows = identity_matrix_rows[rows, :]

identity_matrix_cols = np.eye(number_of_columns)
cols = np.arange(number_of_columns)
np.random.shuffle(cols)
T_cols = identity_matrix_cols[:, cols]

transformed_image = np.matmul(np.matmul(T_rows, img_average_channel), T_cols)
plt.imshow(transformed_image)
plt.show()

Tinv_rows = np.linalg.inv(T_rows)
Tinv_cols = np.linalg.inv(T_cols)
recovered_image = np.matmul(np.matmul(Tinv_rows, transformed_image), Tinv_cols)
plt.imshow(recovered_image)
plt.show()
```
